```python
import os
import vertexai
from vertexai.generative_models import GenerativeModel, GenerationConfig
```

```python
# 1. SET YOUR PROJECT CONFIG HERE
os.environ["VERTEXAI_PROJECT"] = ""
os.environ["VERTEXAI_LOCATION"] = "us-central1"

PROJECT_ID = os.environ["VERTEXAI_PROJECT"]
LOCATION = os.environ["VERTEXAI_LOCATION"]

# 2. Initialize Vertex AI with your specific location
vertexai.init(project=PROJECT_ID, location=LOCATION)

# 3. Use the Gemini 2.5 Flash model
model = GenerativeModel("gemini-2.5-flash")
```

```python
# 4. Generate content (Removed the unsupported thinking_budget config)
response = model.generate_content(
    "Write a one sentence poem about Linux."
)

# 5. Print the final response
print(response.text)
```
